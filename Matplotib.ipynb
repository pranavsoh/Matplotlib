```python
plt.figure#Matlab: Ek nayi blank figure (canvas) create karo jisme hum graphs banayenge.

plt.subplot(1,2,1)#Iska matlab: Hum figure ko 1 row aur 2 columns me divide kar rahe hain.Ye line keh rahi hai: 1st graph left side me dikhana hai.
plt.plot(x,np.sin(x))#x ke against sin(x) ka graph plot karo.Yani yeh sine wave draw karta hai.

plt.subplot(1,2,2)#Ab hum 2nd plot banayenge — yani right side me.Again, figure 1 row & 2 columns ka hai — yeh 2nd position hai.
plt.plot(x,np.cos(x))#x ke against cos(x) ka graph plot karo.Yani yeh cosine wave draw karta hai.

plt.show()#Ab tak jo bhi graphs banaye hain, un sabko screen par display karo.
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)
plt.plot(x,np.cos(x+1))

plt.show()
```

(2, 1, 1) ka matlab hota hai:

2 rows — Figure ko 2 horizontal rows me divide karo.

1 column — Har row me 1 hi column hoga.

1 position — Ab jo subplot banega, wo pehli row ke pehle (aur sirf) column me rahega.

Histogram ka use hum tab karte hain jab hume kisi numerical data ki distribution dekhni ho — yaani data values kaise spread hui hain, kis range me zyada data hai, aur kis range me kam.



```python
#histogram >> continous / numerical data distribution

# Yeh code 1000 random values ka histogram banata hai 
#(normal distribution se), jisme 20 bins hote hain aur red color ('r') ka use kiya gaya hai.

data = np.random.randn(1000) #10 bins
plt.hist(data,color='r',bins=20)
plt.show()
```

A stacked histogram ek aisa histogram hota hai jisme multiple datasets ko ek hi bar me stack karke (upar-upar) dikhaya jaata hai — na ki side-by-side.

✅ Why Use Stacked Histogram? (Hinglish Explanation)
Compare Multiple Groups Together

Jaise boys aur girls ke marks ek hi graph me dekhna ho.

Total Frequency Dikhana

Har bin me total count kya hai, aur kis group ka kitna contribution hai — dono ek saath dikhte hain.

Better Visualization of Proportions

Har segment ka contribution clearly dikhta hai.


```python
#stacked histogram

data1= np.random.normal(0,1,5000)
data2= np.random.normal(0,1,5000)
```

```python
data1
```

```python
data2
```

```python
plt.hist([data1,data2],bins=20 ,color = ['r','b'],label=['dist1','dist2'])
plt.xlabel('Values')
plt.ylabel('frequency')
plt.title('stacked histogram')
plt.show()
```

```python
#pie chart >> TO understand the percent dist of data

langs = ['C','Python','java']
students=[20,100,40]

plt.pie(students,labels = langs ,autopct = '%1.1f%%')
plt.title('pie chart')
plt.show()
```

```python
explode =(0.0,0.1,0.1)

langs = ['C','Python','java']
students=[20,100,40]

plt.pie(students,labels = langs ,autopct = '%1.1f%%',explode=explode ,shadow = True)
plt.title('pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

# Create a area plot
plt.stackplot(years, category1, category2, category3, labels=['Category 1', 'Category 2', 'Category 3'],
              colors=['skyblue', 'lightgreen', 'orange'])

# Add labels and title
plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Area Plot")
plt.legend()

plt.show()
```
